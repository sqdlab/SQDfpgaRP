```python
import scipy as sc
import matplotlib.pyplot as plt
import fxpmath

coefficients = sc.signal.firwin(40, 0.001, window = "hamming", pass_zero="lowpass")
plt.figure(1)
plt.plot(coefficients)
w, h = sc.signal.freqz(coefficients, 1, 512)
print(max(coefficients), min(coefficients))

highest = fxpmath.Fxp(max(coefficients), signed=True, n_word=32, n_frac=32)
lowest = fxpmath.Fxp(min(coefficients), signed=True, n_word=32, n_frac=32)

plt.figure(2)
plt.axhline(y=highest, color='r', linestyle='-')
for idx in range(1, 32):
    approx = fxpmath.Fxp(max(coefficients), signed=True, n_word=idx, n_frac=idx)
    plt.plot(idx, approx, 'bo')
    
plt.figure(3)
plt.axhline(y=lowest, color='r', linestyle='-')
for idx in range(1, 32):
    approx = fxpmath.Fxp(min(coefficients), signed=True, n_word=idx, n_frac=idx)
    plt.plot(idx, approx, 'bo')
```
